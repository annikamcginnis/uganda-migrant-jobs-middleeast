```python
import requests
from bs4 import BeautifulSoup
import time

for company in df_list_dicts:
    if pd.notna(company['job_salary_amount_job_page']):
        #print(company[-2])
        currency = company['job_salary_currency_job_page']
        amount = company['job_salary_amount_job_page']
        url = "https://www.xe.com/currencyconverter/convert/?Amount=" + amount + "&From=" + currency + "&To=USD"
        html = requests.get(url).content 
        soup_doc = BeautifulSoup(html, "html.parser")
        usd_conversion = soup_doc.find(class_='[grid-area:conversion]').find_all('p')[1].next
        #print(usd_conversion)
        company['usd_conversion'] = usd_conversion

df_list_dicts
```

## Geocoding the company locations listed on the companies' pages (sometimes different than other addresses included)

```python
import time 

for company in df_list_dicts:
    url = 'https://maps.googleapis.com/maps/api/geocode/json'
    keys = {'address': company['company_location_companies_page'], 'key': api_key}
    try: 
        r = requests.get(url,params=keys)
        result_dic = r.json()
        if result_dic['status'] == 'OK' and len(result_dic['results']) > 0:
            lat = result_dic['results'][0]['geometry']['location']['lat']
            long = result_dic['results'][0]['geometry']['location']['lng']
            city = result_dic['results'][0]['address_components'][2]['short_name']
            country = result_dic['results'][0]['address_components'][-1]['long_name']
            company['latitude_company_location_companies_page'] = lat
            company['longitude_company_location_companies_page'] = long
            company['city_company_location_companies_page'] = city
            company['country_company_location_companies_page'] = country

        else:
            company['latitude_company_location_companies_page'] = None
            company['longitude_company_location_companies_page'] = None
            company['city_company_location_companies_page'] = None
            company['country_company_location_companies_page'] = None
            print(f"No results for address: {company['company_location_companies_page']} - Status: {result_dic['status']}")
            
    except Exception as e:
        print(f"Error for address: {company['company_location_companies_page']} - {e}")
        company['latitude_company_location_companies_page'] = None
        company['longitude_company_location_companies_page'] = None
        company['city_company_location_companies_page'] = None
        company['country_company_location_companies_page'] = None
        time.sleep(0.5)
```

```python
import time
import requests

for company in df_list_dicts:
    url = 'https://maps.googleapis.com/maps/api/geocode/json'
    keys = {'address': company['company_address_company_page'], 'key': api_key}
    try:
        r = requests.get(url, params=keys)
        result_dic = r.json()
        
        # Initialize default values
        lat = long = city = country = None

        if result_dic['status'] == 'OK' and len(result_dic['results']) > 0:
            result = result_dic['results'][0]
            lat = result['geometry']['location']['lat']
            long = result['geometry']['location']['lng']
            
            # Extract city and country dynamically
            for component in result['address_components']:
                if 'locality' in component['types']:
                    city = component['short_name']
                if 'country' in component['types']:
                    country = component['long_name']
            
            # Assign values to the dictionary
            company['latitude_company_location_company_page'] = lat
            company['longitude_company_location_company_page'] = long
            company['city_company_location_company_page'] = city
            company['country_company_location_company_page'] = country
        else:
            print(f"No results for address: {company['company_address_company_page']} - Status: {result_dic['status']}")
            company['latitude_company_location_company_page'] = None
            company['longitude_company_location_company_page'] = None
            company['city_company_location_company_page'] = None
            company['country_company_location_company_page'] = None

    except Exception as e:
        print(f"Error for address: {company['company_address_company_page']} - {e}")
        company['latitude_company_location_company_page'] = None
        company['longitude_company_location_company_page'] = None
        company['city_company_location_company_page'] = None
        company['country_company_location_company_page'] = None
    finally:
        # Sleep to respect API rate limits
        time.sleep(0.5)

df_list_dicts[0:5]
```
